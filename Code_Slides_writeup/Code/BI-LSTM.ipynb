```python
df['sent']= df['sent'].progress_apply(lambda x: ara.strip_tashkeel(x))
```

```python
from collections import Counter
c=Counter()
for i in tqdm(range(len(df['token']))):
    c.update(df['token'].iloc[i])
c.most_common()
```

```python
vocab_size= len(list(c.keys()))
```

```python
word2int= {j:i for i,j in enumerate(list(c.keys()), 1)}
```

```python
df['token']= df['token'].progress_apply(lambda x: [word2int[i] for i in x])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
```

```python
import pickle
with open('word2ind.pkl', 'wb') as file:
  pickle.dump(word2int, file)
```

```python
sents= pad_sequences(df['token'], 128)
encoder = LabelEncoder()
encoder.fit(df['author'])
encoded_Y = encoder.transform(df['author'])
# convert integers to dummy variables (i.e. one hot encoded)
clas = np_utils.to_categorical(encoded_Y)
```

```python
with open('encoder.pkl', 'wb') as file:
  pickle.dump(encoder, file)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(sents, clas, test_size=0.2, random_state=42)
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.1, random_state=42)
```

```python
#build the LSTM without stopwords removal
import tensorflow.keras as keras
```

```python
# model 1 with stopwords vai_acc= 0.79

model= keras.models.Sequential()
model.add( keras.layers.Embedding(vocab_size, 256, input_length=x_train.shape[-1]) )
model.add( keras.layers.LSTM(512, return_sequences=True) )
model.add( keras.layers.LSTM(64) )
model.add( keras.layers.Dense(df['author'].nunique(), activation='softmax') )
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=3)
```

```python
# model 2 change size vai_acc= 0.81

model= keras.models.Sequential()
model.add( keras.layers.Embedding(vocab_size, 256, input_length=x_train.shape[-1]) )
model.add( keras.layers.LSTM(600, return_sequences=True) )
model.add( keras.layers.LSTM(256) )
model.add( keras.layers.Dense(df['author'].nunique(), activation='softmax') )
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=3)
```

```python
save= keras.callbacks.ModelCheckpoint(filepath='/content/model3.h5',
    monitor='val_accuracy',
    mode='max',
    save_best_only=True)
```

```python
model= keras.models.Sequential()
model.add( keras.layers.Embedding(vocab_size, 300, input_length=x_train.shape[-1]) )
model.add( keras.layers.Bidirectional(keras.layers.LSTM(600, return_sequences=True)) )
model.add( keras.layers.BatchNormalization())
model.add( keras.layers.Bidirectional(keras.layers.LSTM(280)) )
model.add( keras.layers.BatchNormalization())
model.add( keras.layers.Dense(256, activation='relu') )
model.add( keras.layers.BatchNormalization())
model.add( keras.layers.Dense(df['author'].nunique(), activation='softmax') )
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
model.fit(x_train, y_train, validation_data=(x_valid, y_valid), epochs=2)
```

```python
# model 3 bidir

model= keras.models.Sequential()
model.add( keras.layers.Embedding(vocab_size, 350, input_length=x_train.shape[-1]) )
model.add( keras.layers.Bidirectional(keras.layers.LSTM(600, return_sequences=True)) )
model.add( keras.layers.Dropout(0.35) )
model.add( keras.layers.BatchNormalization())
model.add( keras.layers.Bidirectional(keras.layers.LSTM(320)) )
model.add( keras.layers.Dropout(0.3) )
model.add( keras.layers.BatchNormalization())
model.add( keras.layers.Dense(df['author'].nunique(), activation='softmax') )
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=3)
```
